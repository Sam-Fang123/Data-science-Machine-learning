# Task I - Density‚Äêbased Clustering

```python
def _dist(p,q):
    result=((float(p[0])-float(q[0]))**2+(float(p[1])-float(q[1]))**2)**(1/2)
    return result
    # compute and return the euclidean distance
    # put your code here
```

```python
def _eps_neighborhood(p,q,eps):
    if _dist(p,q)<eps:
        return True
    else:
        return False
    # check if the distance between p and q is below eps
    # return True or False
    # put your code here
```

```python
def _region_query(m, point_id, eps):
    n_points = m.shape[1]
    seeds = []
    # find and return all points that belong to eps-neighborhood of point_id
    # put your code here
    for i in range(n_points):
        if i !=point_id:
            if _eps_neighborhood(m[0:2,point_id],m[0:2,i],eps)==True:
                seeds=seeds+[i]
        
    return seeds
```

```python
def _expand_cluster(m, classifications, point_id, cluster_id, eps, min_points):
    # expand the cluster of cluster_id from point_id
    # identify all points belonging to cluster_id
    # update the clustering results in "classification" by assigning cluster_id to each point
    # return True if point_id is a core, False if point_id is not a core
    # write your code below
    seeds=[]
    seeds=_region_query(m,point_id,eps)
    
    for i in seeds:
        classifications[i]=cluster_id
    
    if len(seeds)>=min_points:
        return True
    else:
        return False
```

```python
def dbscan(m, eps, min_points):
    """Implementation of DBSCAN
    You can refer to wikipedia for detailed algorithm: https://en.wikipedia.org/wiki/DBSCAN
    Use Euclidean Distance as the measure
    
    Inputs:
    m - A matrix whose columns are feature vectors
    eps - Maximum distance two points can be to be regionally related
    min_points - The minimum number of points to make a cluster
    
    Outputs:
    An array with either a cluster id number or dbscan.NOISE (None) for each column vector in m
    """
    cluster_id = 1
    n_points = m.shape[1]
    classifications = [UNCLASSIFIED] * n_points
    # the main dbscan algorithm
    # put your code here
    
    for i in range(n_points):
        if classifications[i]!=False:
            continue
        neighbor=[]
        neighbor=_region_query(m,i,eps)
        if len(neighbor)<min_points:
            classifications[i]=NOISE
            continue
        classifications[i]=cluster_id
        
        num_of_n=len(neighbor)
        count=0
        while count<num_of_n:
            n=_expand_cluster(m,classifications,neighbor[count],cluster_id,eps,min_points)
            if n==True:
                neighbor=neighbor+_region_query(m,neighbor[count],eps)
                neighbor=pd.unique(neighbor).tolist()
                num_of_n=len(neighbor)
            count=count+1
            
        cluster_id=cluster_id+1       
                
            
    return classifications
```

```python
def radius(ndarray, r):
    boolean_mask = []
    boolean=[((ndarray**2).sum(axis=0))**(1/2)<r]
    boolean_mask=np.vstack((boolean,boolean))
    # Check if a point is in the given radius
    
    return ndarray[boolean_mask].reshape(2,790)
```

```python
scat_X_Y = radius(scat_X_Y_raw, 1000)
#print(scat_X_Y.shape)
```

```python
%matplotlib inline
plt.figure(figsize = (12, 12))
plt.scatter(scat_X_Y[0,:],scat_X_Y[1,:])
plt.grid()
plt.show()
# Put the "X" and "Y" in the scatter plot function...
# How to make the ÔºÇÊ†ºÁ∑öÔºÇ of X and Y the same length...? Google might help!
```

```python
# Calculates the distance between each point...
dist_sq=np.zeros(790*790)
dist_sq=dist_sq.reshape(790,790)
for i in range(scat_X_Y.shape[1]):
    for j in range(scat_X_Y.shape[1]):
        dist_sq[i,j]=((scat_X_Y[0,i]-scat_X_Y[0,j])**2+(scat_X_Y[1,i]-scat_X_Y[1,j])**2)
        
```

```python
#dist_sq=dist_sq*2
print(dist_sq)
```

```python
# Find the nearest neighbor of each point...
nearest = np.argsort(dist_sq)
```

```python
print(nearest)
```

```python
def nearest_partition(dist_sq, K):
    nearest=np.argsort(dist_sq)
    return nearest[:,0:K+1]
# Use a numpy method(function) to return the "K Nearest Neighbor"
```

```python
plt.figure(figsize = (12, 12))
plt.scatter(scat_X_Y[0,:],scat_X_Y[1,:])
plt.grid()
    # Put the "X" and "Y" in the scatter plot function...

# draw lines from each point to its three nearest neighbors
K = 3
k_nearest=nearest_partition(dist_sq,K)
for i in range(scat_X_Y.shape[1]):
    for n in k_nearest[i]:
        plt.plot([scat_X_Y[0,i],scat_X_Y[0,n]],[scat_X_Y[1,i],scat_X_Y[1,n]],color='k',linewidth=1)
# You might need loops to generate the plot below...

# How to make the ÔºÇÊ†ºÁ∑öÔºÇ of X and Y the same length...? Google might help!
plt.show()
```

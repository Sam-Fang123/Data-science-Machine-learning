```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

def dis(p1_x,p1_y,p2_x,p2_y):         #function
    distance=(p1_x-p2_x)**2+(p1_y-p2_y)**2
    return distance**(1/2)

# Read data file
file_to_load = "cdata.csv"

# Read purchasing file and store into pandas data frame
cdata = pd.read_csv(file_to_load)
index=np.random.randint(0,len(cdata),size=4)    #40~cdataindex
c=[]                                            #c
for j in index:
    c.append([cdata.iloc[j]['x'],cdata.iloc[j]['y']])   #assign
pp=0
while True:    
    cluster=[]
    for i in range(len(cdata)):
        d=[]                                    #d
        for j in range(len(c)):
          d.append(dis(cdata.iloc[i]['x'],cdata.iloc[i]['y'],c[j][0],c[j][1]))   #ddis function
        cluster=cluster+[d.index(min(d))]       #dassign
    cdata['new cluster']=cluster
    temp_c=[]
    for k in range(4):                          #temp_c
        temp_c.append([cdata[cdata['new cluster']==k]['x'].mean(),cdata[cdata['new cluster']==k]['y'].mean()])
    
    if c==temp_c:                               #ctemp_c
        break
    else:
        c=temp_c
    color=['b.','g.','y.','r.']                #
    label1=['A','B','C','D']                   #
    for i in range(len(color)):
        plt.plot(cdata[cdata['new cluster']==i]['x'],cdata[cdata['new cluster']==i]['y'],color[i],label=label1[i])  #
        plt.plot(temp_c[i][0],temp_c[i][1],'k*',label='center')
    plt.ylabel("y") # y label
    plt.xlabel("x") # x label
    plt.legend(loc = "best", fontsize=10)
    plt.show()
```

```python

index2=[]
for i in range(2,11):
    index2.append(list(np.random.randint(0,len(cdata),size=i)))   #assgin2~10index

SSE_LIST=[]    #sse

for i in range(len(index2)):        #2~10
    c2=[]     #c2
    for j in index2[i]:
        c2.append([cdata.iloc[j]['x'],cdata.iloc[j]['y']]) 
    while True:    #while
        cluster=[]
        for i in range(len(cdata)):
            d=[]
            for j in range(len(c2)):
                d.append(dis(cdata.iloc[i]['x'],cdata.iloc[i]['y'],c2[j][0],c2[j][1]))
            cluster=cluster+[d.index(min(d))]
        cdata['new cluster']=cluster
        temp_c=[]
        for k in range(len(index2)):
            temp_c.append([cdata[cdata['new cluster']==k]['x'].mean(),cdata[cdata['new cluster']==k]['y'].mean()])
    
        if c2==temp_c:
            break
        else:
            c2=temp_c

    SSE=0
    for i in range(len(c2)):      #SSE
        SSE=SSE+((cdata[cdata['new cluster']==i]['x']-c2[i][0])**2).sum()+((cdata[cdata['new cluster']==i]['y']-c2[i][1])**2).sum()
    SSE_LIST.append(SSE)
    #print(SSE)
plt.plot(range(2,11),SSE_LIST)    #
plt.ylabel("Sum of square error") # y label
plt.xlabel("Number of k") # x label
plt.show()
```
